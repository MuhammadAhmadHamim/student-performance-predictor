```python
df['Gender'] = df['Gender'].str.title()
# Converting all the String values inside the Gender column into a title(Initial Capitalized) 
# and overwriting the result onto the same Gender column
```

```python
df['Name'] = df['Name'].str.title()
# Converting all the String values inside the Name column into a title(Initial Capitalized) 
# and overwriting the result onto the same Name column
```

```python
df['Department'] = df['Department'].str.upper().str.strip()
# Converting all the abreviations of department names into upper case letters using the upper() method and 
# removing the surplus spaces using the strip() method. Both methods belong to String dataType
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors= 'coerce') 
# Converts the whole column to numeric, errors = coerce converts the invalid values into Null(NaN) 
# and overwriting the result onto the same Age column

df['Age'] = df['Age'].fillna(df['Age'].median())
# Filling all the null values of Age column with the median of the column.
# Why Median and not mean?
# Because the column has a lotta values missing(NUll) and as it is demand of the project, so we replace NaN with median not mean.
```

```python
df['Attendance_%'] = df['Attendance_%'].mask((df['Attendance_%'] > 100) | (df['Attendance_%'] < 0), np.nan)
# Using the .mask() method to replace all the values which satisfy any(OR operator symbol pipe(|)) of the two conditions with NaN(NULL)

df['Attendance_%'] = df['Attendance_%'].fillna(df['Attendance_%'].median())
# Filling the null values with the median
```

```python
df['Assignments'] = df['Assignments'].fillna(df['Assignments'].median())
# Filling the null values with the median
```

```python
df['Study_Hours_Daily'] = df['Study_Hours_Daily'].fillna(df['Study_Hours_Daily'].median())
# Filling the null values with the median
```

```python
df['Midterm'] = df['Midterm'].fillna(df['Midterm'].median())
# Filling the null values with the median
```

```python
df['Final_Score'] = df['Final_Score'].mask((df['Final_Score'] > 100), np.nan)
# Using the .mask() method to replace all the values which satisfy condition with NaN(NULL)

df['Final_Score'] = df['Final_Score'].fillna(df['Final_Score'].median())
# Filling the null values with the median
```

```python
df.drop_duplicates(inplace = True) # Dropping all the duplicates 
```

```python
print(df.isnull().sum()) # Final check to confirm the existence of a null value throughout the dataFrame
```

```python
df.to_excel('../data/cleaned_student_performance_dataset.xlsx', index=False) # Exporting the cleaned data into an excel file
```
