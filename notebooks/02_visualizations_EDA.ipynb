```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_cleaned = pd.read_excel('../data/cleaned_student_performance_dataset.xlsx') # Let's load the cleaned data from our exported file
```

```python
# Chart1:

plt.figure(figsize=(8,5))

sns.histplot(df_cleaned['Final_Score'])

plt.title('Distribution of Final Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.show()
```

```python
# This chart clearly shows the distribution of different values of final scores spread across different students.
# We can clearly see that most students(almost 80) have their final score equal to 40-42.5.
# While the range 57.5-60.0 has the least amount of students(barely 5). 
# This graph is a visual representation of final scores of students grouped by final score values, 
# with the bars showing the number(frequency) of students
```

```python
# Chart2:

plt.figure(figsize=(8,5))

sns.scatterplot(x=df_cleaned['Attendance_%'], y=df_cleaned['Final_Score'])

plt.title('Attendance_% vs Final_Score')
plt.xlabel('Attendance(%)')
plt.ylabel('Final Scores')

plt.show()
```

```python
# This scatter plot shows the final scores of students in correspondance to their attendance percentage.
# We can see that as the attendance_% of a student increases it's grade also improves. 
# The students with the highest grades have their attendance(%) in between 90-100 while,
# usually, students with higher attendance percentage have higher grades and vice versa.
```

```python
# Chart3:

plt.figure(figsize=(8,5))

sns.scatterplot(x=df_cleaned['Study_Hours_Daily'], y=df_cleaned['Final_Score'])

plt.title('Daily Study Hours vs Final_Score')
plt.xlabel('Daily Study Hours')
plt.ylabel('Final Scores')

plt.show()
```

```python
# This scatter plot shows the final scores of students in correspondance to their daily study hours.
# We can see that greater the number of study hours of a student, higher is it's grade(with a few exceptions). 
# There's a gradual increase in the grades of students who have 7 hours of daily study all the way to 60.0.
# While, there are students having a 40.0 grade in almost all the study hours values.
```

```python
# Chart4:

plt.figure(figsize=(8,5))

sns.scatterplot(x=df_cleaned['Midterm'], y=df_cleaned['Final_Score'])

plt.title('Midterm_Score vs Final_Score')
plt.xlabel('Midterm')
plt.ylabel('Final Scores')

plt.show()
```

```python
# This scatter plot shows the final scores of students in correspondance to their midterm marks.
# We can see that higher the marks in midterm, higher is the chance to have better final score and vice versa. 
# There's no student who scored the lowest grade while having the highest marks possible(>90) in midterm.
# While, there are students having a 40.0 grade who, mostly, choked in midterm with scores less than 75.
```

```python
# Chart5

plt.figure(figsize=(8,6))

sns.heatmap(df_cleaned.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Correlation Between Numerical Columns")

plt.show()
```

```python
# This chart describes the heatMap of all the numerical columns of the given dataFrame.
# This HeatMap clearly shows that most numerical columns in the dataFrame are independent of each other(with few exceptions).
# The Final_Score variable depends to some extent on the values of Study_Hours_Daily, Attendance_%, Assignments and Quizzes all the while
# it highly depends on the scores of Midterm column which is exactly what we eeked out during our Midterm vs Final_Score scatter plot 
# that higher the marks in Midterms, higher is the chance to score good Final_Score marks.
# There are other minor relations like between Attendance and Quizzes, Study_Hours_Daily and Quizzes, 
# all the while Student_ID and Age are completely independent variables with no relation(correlation < 0) with any of the other columns.
```

```python
# Chart6:

plt.figure(figsize=(10,5))

sns.boxplot(x=df_cleaned['Department'], y=df_cleaned['Final_Score'])

plt.title('Score Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Final Score')

plt.show()
```

```python
# This boxplot demonstrates the distribution of final scores of student labelled as per their departments.
# Every department has the box which is where most values lie, along with the median line inside it, 
# the position of median line indicates the consistency through the values,
# some whiskers to show the range of values and only BBA department has an outlier indicating one unusual/extreme record(maybe 60?).
```

```python
# Chart7:

plt.figure(figsize=(8,5))

sns.countplot(df_cleaned['Department'], order= df_cleaned['Department'].value_counts().index)

plt.title('Student Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Number of Students')

plt.show()
```

```python
# This countplot shows the number of students per department.
# The most populated department is EE(Electrical Engineering) while the least populated department is ME(Mechanical Engineering)
```
